```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
```

```python
data = load_breast_cancer()
X = data.data
y= data.target
```

```python
def pred(X,weights):
    X = np.append(X,1)
    j= X.T.dot(weights)
    return 1 if j[0]>=0 else 0
    
```

```python
def train(X,y,iterations):
    weights = np.random.randn(X.shape[1]+1,1)
    for x in range(iterations):
        _ = np.random.randint(0,X.shape[0])
        point = X[_]
        prediction = pred(point,weights)
        if y[_] == 1 and prediction == 0:
            weights = weights+np.append(X[_],1).reshape(X.shape[1]+1,1)
        if y[_] == 0  and prediction == 1:
            weights = weights-np.append(X[_],1).reshape(X.shape[1]+1,1)
        if x %1000 ==0 :
            predictions = [pred(j,weights) for j in X ]
            print(accuracy_score(predictions,y))
    return weights
    
            
        
```

```python
weights = train(X,y,1000000)
```
